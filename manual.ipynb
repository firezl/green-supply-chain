```python
sample_data = survey_data[survey_data["education"] == "D.大学本科"][survey_data["id"] != 28][survey_data["id"] != 43][survey_data["id"] != 19][survey_data["id"] != 2][survey_data["id"] != 0][survey_data["id"] != 30]
```

```python
# 保存sample_data
sample_data.to_excel(data_path + "sample_data.xlsx")
```

```python
N = sample_data.shape[0]
K = sample_data.shape[1] - 1
```

```python
N
```

```python
K
```

```python
item_variances = sample_data.var(ddof=1)
```

```python
cov_sum = 0
for i in range(K):
    for j in range(i+1, K):
        cov_sum += sample_data.iloc[:,i].cov(sample_data.iloc[:,j])
```

```python
cronbach_alpha = K/(K-1)*(1 - cov_sum/(item_variances.sum() + cov_sum))
```
